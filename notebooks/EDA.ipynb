# Análise e Modelagem para Manutenção **Preditiva**

***Obtenção e Pré-processamento dos Dados***

A base de dados bootcamp_train.csv foi utilizada para este projeto de manutenção preditiva. A análise inicial revelou a necessidade de um robusto pré-processamento para garantir a qualidade dos dados e a eficácia do modelo.


***Tratamentos Realizados***

**Limpeza e Conversão de Colunas Binárias:** As colunas que representam falhas binárias (como FDF, FDC, FP, FTE, FA e falha_maquina) apresentavam inconsistências. Valores como 'N' ou 'não' foram padronizados e convertidos para o formato booleano (True ou False), facilitando a análise e o uso posterior em modelos.

**Tratamento de Valores Nulos:** A coluna desgaste_da_ferramenta continha valores nulos. Para preencher essas lacunas, foi utilizado o SimpleImputer com a estratégia de imputação pela média. Isso garantiu que o volume de dados fosse preservado sem introduzir vieses significativos.

**Remoção de Colunas Constantes:** A coluna umidade_relativa foi identificada como constante, ou seja, todos os seus valores eram idênticos. Colunas com essa característica não fornecem informações relevantes para a modelagem e, portanto, foram removidas do dataset.

**Interpretação de Valores Negativos:** A coluna temperatura_ar apresentou valores negativos, o que, a princípio, pode parecer um erro. No entanto, após uma análise cuidadosa do contexto, foi interpretado que esses valores poderiam representar anomalias ou falhas de sensores, sendo tratado como um dado válido para o modelo. No caso do modelo XGBoost, ele é robusto a essas variações, o que evita a necessidade de remoção ou imputação de forma drástica.

**Transformação para Problema de Classificação Multiclasse:** As múltiplas colunas binárias de falhas (FDF, FDC, FP, FTE, FA) foram consolidadas em uma única coluna chamada classe_defeito. Essa abordagem converteu o problema de uma classificação multirrótulo para uma classificação multiclasse, simplificando a modelagem.

**Criação de Novas Features:** Foram criadas novas features combinadas, como razões e produtos de variáveis existentes (torque e velocidade_rotacional), para capturar interações entre as variáveis e enriquecer a representatividade dos dados.

**Remoção de Outliers:** Utilizando o método do Interquartile Range (IQR), foram identificados e removidos outliers para cada classe de falha, garantindo que o modelo não fosse excessivamente influenciado por pontos de dados atípicos.

**Modelagem Preditiva**

Tipo de Modelagem e Modelo Utilizado

O problema foi abordado como uma tarefa de classificação multiclasse. O modelo escolhido foi o XGBoostClassifier, conhecido por sua alta performance e robustez. O modelo foi integrado em um pipeline completo para automatizar o fluxo de trabalho.

Pipeline de Pré-processamento: Um ColumnTransformer foi utilizado para aplicar transformações específicas a colunas numéricas (por exemplo, padronização) e categóricas (como One-Hot Encoding).

Balanceamento de Classes: Para lidar com o desbalanceamento de classes, a técnica SMOTE (Synthetic Minority Over-sampling Technique) foi aplicada no conjunto de treino. Isso ajudou a criar amostras sintéticas para as classes minoritárias, nivelando a distribuição e melhorando o desempenho do modelo nessas classes.

Otimização de Hiperparâmetros: Os hiperparâmetros do XGBoost (como max_depth, learning_rate e reg_alpha) foram ajustados para otimizar o desempenho e evitar overfitting.

**Métricas de Avaliação**

As seguintes métricas foram utilizadas para avaliar o desempenho do modelo:

F1-Score por Classe (Micro Análise): Avalia a precisão e a revocação para cada classe individualmente.

F1-Macro (Macro Análise): Considera a média do F1-Score de cada classe. Essa métrica é ideal para conjuntos de dados desbalanceados, pois dá a mesma importância para as classes minoritárias e majoritárias, fornecendo uma visão mais realista do desempenho global do modelo.

Matriz de Confusão: Uma visualização que detalha os acertos e erros do modelo, permitindo identificar quais classes estão sendo confundidas.

A validação foi realizada usando a validação cruzada (cross-validation) com 5 folds e a métrica f1_macro, garantindo que o desempenho do modelo fosse consistente e não dependesse de uma partição específica dos dados.

**Desbalanceamento de Classes**

O dataset original apresentou um forte desbalanceamento, com algumas classes de falha sendo muito mais raras do que outras:

A classe falha_outros era altamente dominante.

Classes como falha_5 e falha_4 tinham pouquíssimas amostras.

Para mitigar esse problema, as seguintes soluções foram adotadas:

Aplicação de SMOTE: O SMOTE foi fundamental para balancear as classes no conjunto de treino, fornecendo ao modelo dados mais representativos das classes minoritárias.

Tuning de Hiperparâmetros do XGBoost: O ajuste fino do modelo ajudou a otimizar sua capacidade de generalização, mesmo com as classes desbalanceadas.

Uso de Métricas Adequadas: A avaliação com o F1-Macro garantiu que o desempenho nas classes raras fosse considerado, evitando que o modelo simplesmente se concentrasse na classe majoritária.

**CÓPIA DO DATAFRAME - SEM IMPUTAÇÃO**

```python

# Copiando o DataFrame original (sem imputação)
df_final = df.copy()

# Criando uma cópia para imputação
df_imputado = df.copy()

# Salvando os nulos antes da imputação
nulos_antes = df_imputado.isnull().sum()

# Imputando valores faltantes em colunas numéricas com a média
for col in df_imputado.select_dtypes(include=['int64', 'float64']).columns:
    df_imputado[col].fillna(df_imputado[col].mean(), inplace=True)

# Imputando valores faltantes em colunas categóricas com a moda
for col in df_imputado.select_dtypes(include=['object']).columns:
    df_imputado[col].fillna(df_imputado[col].mode()[0], inplace=True)

# Salvando os nulos depois da imputação
nulos_depois = df_imputado.isnull().sum()

# Comparando antes e depois
verificacao = pd.DataFrame({
    'Nulos Antes': nulos_antes,
    'Nulos Depois': nulos_depois,
    'Total Preenchido': nulos_antes - nulos_depois
})

# Calculando porcentagem de nulos preenchidos
verificacao['% Preenchido'] = (verificacao['Total Preenchido'] / df.shape[0]) * 100

# Adicionando linha de total
linha_total = pd.DataFrame(verificacao.sum(numeric_only=True)).T
linha_total.index = ['TOTAL']
linha_total['% Preenchido'] = (linha_total['Total Preenchido'] / (df.shape[0] * df.shape[1])) * 100

# Concatenando
verificacao = pd.concat([verificacao, linha_total])

print("Resumo da imputação:")
print(verificacao)
```

```python
cols_to_process = list(set(colunas_falha + ['falha_maquina']).intersection(df.columns))

for col in cols_to_process:
    df[col] = df[col].astype(str).str.strip().str.lower()


```

```python
# Aplicando lógica de exclusão mútua para os hifens ('-')
for col in colunas_falha:
    outras = [c for c in colunas_falha if c != col]  # as outras falhas
    # Se a coluna atual tiver '-', substitui baseado nas outras (se alguma estiver ativa = '1')
    df.loc[df[col] == '-', col] = df[outras].apply(
        lambda row: '0' if '1' in row.values else '1', axis=1
    )
```

```python
# Selecionando colunas de falha específicas (exceto 'falha_maquina')
colunas_falha = [col for col in df.columns if 'falha' in col.lower() and col != 'falha_maquina']

# Função para normalizar cada valor
def normaliza_valor(x):
    x = str(x).strip().lower()
    if x in ['1', 'true', 'sim']:
        return 1
    elif x in ['0', 'false', 'não', 'nao']:
        return 0
    elif x == '-':
        return pd.NA
    else:
        return pd.NA  # qualquer outro valor estranho vira NA

# Aplicando a normalização
for col in colunas_falha:
    df[col] = df[col].apply(normaliza_valor).astype('Int64')

# Preencher valores faltantes com base em uma coluna de referência (primeira coluna)
referencia = colunas_falha[0]
for col in colunas_falha[1:]:
    df.loc[df[col].isna(), col] = df[referencia].map({1: 0, 0: 1}).astype('Int64')

# Checando inconsistências: duas falhas específicas iguais a 1 ou 0
for i, col1 in enumerate(colunas_falha):
    for col2 in colunas_falha[i+1:]:
        print(f"{col1} = 1 e {col2} = 1?",
              df[(df[col1] == 1) & (df[col2] == 1)].shape[0])
        print(f"{col1} = 0 e {col2} = 0?",
              df[(df[col1] == 0) & (df[col2] == 0)].shape[0])

# Consultando registros com duas falhas iguais a 0
inconsistentes = df[(df[colunas_falha[0]] == 0) & (df[colunas_falha[1]] == 0)]
print("Registros com duas falhas iguais a 0:")
print(inconsistentes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------- Passo 1: Variáveis para análise ----------
variaveis = ['temperatura_ar', 'temperatura_processo', 'velocidade_rotacional', 'desgaste_da_ferramenta']

# ---------- Passo 2: Criar tabela de contagem e proporção ----------
resultado = pd.DataFrame(columns=['Variável', 'Negativo', 'Positivo', 'Proporção Negativo (%)', 'Proporção Positivo (%)'])

for col in variaveis:
    neg = (df[col] < 0).sum()
    pos = (df[col] >= 0).sum()
    total = neg + pos
    prop_neg = neg / total * 100 if total > 0 else 0
    prop_pos = pos / total * 100 if total > 0 else 0
    resultado = pd.concat([resultado, pd.DataFrame([[col, neg, pos, prop_neg, prop_pos]],
                                                    columns=['Variável', 'Negativo', 'Positivo',
                                                             'Proporção Negativo (%)', 'Proporção Positivo (%)'])],
                           ignore_index=True)

# Exibir tabela
display(resultado)

```

```python
# ----------  Gráfico 1 - Valores absolutos ----------
plt.figure(figsize=(8,6))
plt.bar(resultado['Variável'], resultado['Negativo'], label='Negativo', color='red')
plt.bar(resultado['Variável'], resultado['Positivo'], bottom=resultado['Negativo'], label='Positivo', color='blue')

# Anotando valores nas barras
for i, var in enumerate(resultado['Variável']):
    neg = resultado.loc[i, 'Negativo']
    pos = resultado.loc[i, 'Positivo']
    plt.text(i, neg / 2, f"{neg}", ha='center', va='center', color='white', fontweight='bold')
    plt.text(i, neg + pos / 2, f"{pos}", ha='center', va='center', color='black', fontweight='bold')

plt.ylabel('Número de registros')
plt.title('Distribuição de Registros por Sinal (Valores Absolutos)')
plt.xticks(rotation=25)
plt.legend(title='Sinal')
plt.tight_layout()
plt.show()


```

```python
# ----------  Gráfico 2 - Proporções (%) ----------
plt.figure(figsize=(8,6))
plt.bar(resultado['Variável'], resultado['Proporção Negativo (%)'], label='Negativo', color='red')
plt.bar(resultado['Variável'], resultado['Proporção Positivo (%)'], bottom=resultado['Proporção Negativo (%)'], label='Positivo', color='blue')

# Anotando proporções nas barras
for i, var in enumerate(resultado['Variável']):
    neg = resultado.loc[i, 'Proporção Negativo (%)']
    pos = resultado.loc[i, 'Proporção Positivo (%)']
    plt.text(i, neg / 2, f"{neg:.1f}%", ha='center', va='center', color='white', fontweight='bold')
    plt.text(i, neg + pos / 2, f"{pos:.1f}%", ha='center', va='center', color='black', fontweight='bold')

plt.ylabel('Proporção (%)')
plt.title('Distribuição de Registros por Sinal (Proporções)')
plt.xticks(rotation=25)
plt.legend(title='Sinal')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

**INTERPRETAÇÃO** - **GRÁFICO DE PROPORÇÕES**

A barra é majoritariamente azul (positivo) → a variável tem mais valores positivos.

As proporções não são equilibradas (positivo/negativo) → a variável com valores positivo é supeiror a variável com valores negativos.


Essa análise é útil para entender o sinal predominante em cada variável.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ----------  Colunas de falha simuladas ----------
np.random.seed(42)
df['FDF'] = np.random.randint(0,2,size=len(df))
df['FDC'] = np.random.randint(0,2,size=len(df))
df['FP']  = np.random.randint(0,2,size=len(df))
df['FTE'] = np.random.randint(0,2,size=len(df))
df['FA']  = np.random.randint(0,2,size=len(df))

falhas = ['FDF','FDC','FP','FTE','FA']
variaveis = ['temperatura_ar', 'temperatura_processo', 'velocidade_rotacional', 'desgaste_da_ferramenta']

# ---------- Passo 1: Preparar dados para gráfico ----------
dados_grafico = {}
for col in variaveis:
    dados_grafico[col] = {
        'Negativo': df[df[col]<0][falhas].sum(),
        'Positivo': df[df[col]>=0][falhas].sum()
    }

# ---------- Passo 2: Criar gráfico ----------
fig, ax = plt.subplots(figsize=(8,6))
bar_width = 0.35
x = np.arange(len(variaveis))
colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00']  # cores por falha

for i, col in enumerate(variaveis):
    bottom_neg = 0
    bottom_pos = 0
    for j, falha in enumerate(falhas):
        # Barra Negativa
        val_neg = dados_grafico[col]['Negativo'][falha]
        ax.bar(x[i] - bar_width/2, val_neg, bottom=bottom_neg, color=colors[j])
        bottom_neg += val_neg
        # Barra Positiva
        val_pos = dados_grafico[col]['Positivo'][falha]
        ax.bar(x[i] + bar_width/2, val_pos, bottom=bottom_pos, color=colors[j], alpha=0.6)
        bottom_pos += val_pos

ax.set_xticks(x)
ax.set_xticklabels(variaveis, rotation=25)
ax.set_ylabel('Número de Falhas')
ax.set_title('Distribuição de Falhas por Variável (Negativo vs Positivo)')
# Legenda: cada cor = falha
ax.legend(falhas, title='Tipo de Falha', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pointbiserialr

# ========================
# 1. Normalizar nomes das colunas (remover descrições entre parênteses)
# ========================
df.columns = df.columns.str.replace(r"\(.*\)", "", regex=True).str.strip()

# Detectar colunas de falhas
falhas = [col for col in df.columns if 'falha' in col.lower() or col.upper() in ['FDF','FDC','FP','FTE','FA']]
print("Colunas de falhas detectadas:", falhas)

# Garantir que são numéricas
df[falhas] = df[falhas].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)

```

```python

# ========================
# 3. Analisar influência de umidade e temperatura nas falhas
# ========================
variaveis_ambiente = [col for col in ['umidade_relativa', 'temperatura_ar', 'temperatura_processo'] if col in df.columns]

# Médias por falha
if 'falha_maquina' in df.columns:
    for var in variaveis_ambiente:
        plt.figure(figsize=(10, 5))
        sns.boxplot(x='falha_maquina', y=var, data=df)
        plt.title(f'{var} vs falha_maquina')
        plt.show()
```

```python
# ========================
# 5. Correlação (Point-Biserial)
# ========================
if 'falha_maquina' in df.columns:
    for var in variaveis_ambiente:
        corr, p_value = pointbiserialr(df['falha_maquina'], df[var])
        print(f"{var}: Correlação={corr:.4f}, p-valor={p_value:.4f}")
```

#**ANÁLISE MULTIVARIADA - PCA**

```python
colunas_falha = [
    'falha_maquina',
    'FDF (Falha Desgaste Ferramenta)',
    'FDC (Falha Dissipacao Calor)',
    'FP (Falha Potencia)',
    'FTE (Falha Tensao Excessiva)',
    'FA (Falha Aleatoria)'
]

# Ajuste se necessário
colunas_falha = [col.strip() for col in colunas_falha if col.strip() in df.columns]
```

```python
# Seleciona colunas numéricas do DataFrame
variaveis_numericas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Variáveis numéricas:", variaveis_numericas)
```

```python
variaveis_continuas = [col for col in variaveis_numericas if col in df.columns and col not in colunas_falha]
print("Variáveis contínuas usadas:", variaveis_continuas)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 1️⃣ Ajustar colunas de falha que realmente existem
colunas_falha_existentes = [col for col in colunas_falha if col in df.columns]

# 2️⃣ Selecionar variáveis contínuas existentes
variaveis_continuas = [col for col in variaveis_numericas if col in df.columns and col not in colunas_falha_existentes]

# 3️⃣ Padronização
X_scaled = StandardScaler().fit_transform(df[variaveis_continuas])

# 4️⃣ PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# DataFrame PCA
df_pca = pd.DataFrame(X_pca, columns=['PC1','PC2'])
df_pca = pd.concat([df_pca, df[colunas_falha_existentes]], axis=1)

```

```python
# 5️⃣ Plotando cada falha com densidade e biplot
plt.figure(figsize=(6,6))

colors = sns.color_palette('Set1', len(colunas_falha_existentes))

for i, falha in enumerate(colunas_falha_existentes):
    subset = df_pca[df_pca[falha]==1]
    sns.kdeplot(
        x=subset['PC1'], y=subset['PC2'],
        fill=True, alpha=0.3, color=colors[i],
        label=falha, thresh=0.05
    )
    plt.scatter(subset['PC1'], subset['PC2'], color=colors[i], alpha=0.6, s=30)

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# ---------------------------
# 1. Descobrir coluna de categoria (falha/tipo/classe) ou criar uma genérica
# ---------------------------
colunas_df = df.columns.tolist()
colunas_candidatas = [c for c in colunas_df if any(x in c.lower() for x in ['falha', 'tipo', 'classe', 'defeito'])]

if len(colunas_candidatas) == 0:
    print("Nenhuma coluna de classe encontrada. Criando coluna genérica...")
    df['grupo'] = 'Categoria Única'
    coluna_categoria = 'grupo'
else:
    coluna_categoria = colunas_candidatas[0]
    print(f"Usando a coluna: {coluna_categoria}")

# ---------------------------
# 2. Adicionar coluna ao df_pca
# ---------------------------
if 'grupo_plot' not in df_pca.columns:
    df_pca['grupo_plot'] = df[coluna_categoria].values

# ---------------------------
# 3. Gerar gráfico PCA com Biplot
# ---------------------------
escala = 5  # escala para os vetores

plt.figure(figsize=(8,8))

# Plot dos pontos PCA coloridos por tipo
for tipo in df_pca['grupo_plot'].unique():
    idx = df_pca['grupo_plot'] == tipo
    plt.scatter(df_pca.loc[idx, 'PC1'], df_pca.loc[idx, 'PC2'], label=tipo, alpha=0.7)

# Vetores do biplot
loadings = pd.DataFrame(pca.components_.T, index=variaveis_continuas, columns=['PC1','PC2'])
for i, var in enumerate(variaveis_continuas):
    plt.arrow(0, 0, loadings.PC1[i]*escala, loadings.PC2[i]*escala,
              color='red', alpha=0.8, head_width=0.15, head_length=0.25)
    plt.text(loadings.PC1[i]*escala*1.1, loadings.PC2[i]*escala*1.1, var,
             color='red', fontsize=10, ha='center', va='center',
             bbox=dict(facecolor='white', alpha=0.6, edgecolor='none', pad=1))

# Layout
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variância)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variância)')
plt.title('PCA Avançado - Biplot com Tipo de produto/máquina')
plt.grid(True)
plt.legend(title='Tipo', bbox_to_anchor=(1.05,1), loc=2)
plt.tight_layout()
plt.show()
```

```python
#  Vetores das variáveis contínuas (biplot)
loadings = pd.DataFrame(pca.components_.T, index=variaveis_continuas, columns=['PC1','PC2'])
for i, var in enumerate(variaveis_continuas):
    plt.arrow(0, 0, loadings.PC1[i]*5, loadings.PC2[i]*5, color='red', alpha=0.7, head_width=0.2)
    plt.text(loadings.PC1[i]*5.2, loadings.PC2[i]*5.2, var, fontsize=10)

plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variância)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variância)')
plt.title('PCA Avançado - Multiplas Falhas com Contornos e Biplot')
plt.grid(True)
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.legend(title='Falhas', bbox_to_anchor=(1.05,1), loc=2)
plt.tight_layout()
plt.show()


```
